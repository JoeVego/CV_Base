```python
import numpy as np
```

```python
pred = np.array([3, -0.5, 2, 8])
true = np.array([2.5, 0.0, 2, 8])

true1 = np.array([[0.5, 1],
           [-1, 1],
           [7, -6]])
pred1 = np.array([[0.5, 2],
           [-1, 2],
           [-7, -5]])
```

```python
a = np.zeros((2, 3, 4, 5))
```

```python
def get_dims(np_array):
    #возвращает 1 если вектор и 2 если 2мерный вектор и тд.
    return len(np.shape(np_array))
```

```python
def accuracy(pred, true):
    dims = get_dims(true)
    result = 0
    
    if dims == 1:
        idx = 0
        vec_size = true.size
        counter = 0
        while idx < vec_size:
            if true[idx] == pred[idx]:
                counter += 1
                
            idx += 1
        result = counter / vec_size
        
    if dims == 2:
        counter = 0
        rows = len(true)
        cols = len(true[0])
        counter_f = 0 
        counter_s = 0
        
        while counter_f < rows:
            while counter_s < cols:
                if true[counter_f][counter_s] == pred[counter_f][counter_s]:
                    counter += 1
                    
                counter_s += 1
            counter_s = 0
            counter_f += 1
            
        result = counter / true.size
            
            
    return result
```

```python
print(accuracy(pred, true))
```

```python
print(accuracy(pred1, true1))
```

```python
print(error_rate(pred, true))
```

```python
print(error_rate(pred1, true1))
```
