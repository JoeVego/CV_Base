z-score normalization

```python
import math
```

```python
features = [4, 6, 7, 34, 32, 85, 54, 40, 9, 90]
```

```python
def mean_deviation(features_list):
    feat_sum =  0
    
    for feature_val in features_list:
        feat_sum += feature_val
    
    mean_val = feat_sum / len(features_list)
    
    return mean_val
```

```python
print(mean_deviation(features))
```

```python
def standard_deviation(features_list):
    mean_dev_val = mean_deviation(features)
    diff_val =  0
    
    for feature_val in features_list:
        temp_val = feature_val - mean_dev_val
        temp_val = temp_val * temp_val
        diff_val += temp_val
    
    diff_val = diff_val / len(features_list)
    
    return math.sqrt(diff_val)
```

```python
print(standard_deviation(features))
```

```python
def scaling(features_list):
    mean_dev_val = mean_deviation(features)
    standart_dev_val = standard_deviation(features)
    temp_list  =  []
    result = []
    
    for feature_val in features_list:
        temp = feature_val - mean_dev_val
        temp = temp / standart_dev_val
        temp_list.append(temp)
    
    for val in temp_list:
        result.append(round(val, 3))
        
    return result
```

```python
res_list = scaling(features)
for val in res_list:
    print(round(val, 3))
```
