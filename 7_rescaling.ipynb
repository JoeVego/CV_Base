rescaling(или min-max нормализации).

```python
import math
```

```python
features = [4.0, 6.0, 7.0, 34.0, 32.0, 85.0, 54.0, 40.0, 9.0, 90.0]
```

```python
def get_min(somelist):
    min_val = somelist[0]
    
    for val in somelist:
        if val < min_val :
            min_val = val
    return min_val
```

```python
def get_max(somelist):
    max_val = somelist[0]
    
    for val in somelist:
        if val > max_val :
            max_val = val
    return max_val
```

```python
def rescaling(features_list):
    min_val = get_min(features_list)
    max_val = get_max(features_list)
    output_list = []
    
    for val in features_list:
        val = (val - min_val) / (max_val - min_val)
        output_list.append(round(val,3))
    
    return output_list
```

```python
print(rescaling(features))
```
