```python
vec_1 = [-4.6, 3.0, 90.4, -65.67, 0.7, -45.1, 76.0]
vec_2 = [-5.6, 67.4, 50.2, -88.1, -33.1, -22.0, 33.3]
```

```python
def mult(vec1, vec2):
    list_len = len(vec1)
    counter = 0
    sum_val = 0
    
    if list_len != len(vec2):
        raise Exception("Vecs not have same length !")
    
    while counter < list_len:
        sum_val = (vec1[counter] * vec2[counter]) + sum_val
        counter += 1
    
    return sum_val
```

```python
print(mult(vec_1, vec_2))
```

```python
vec_4 = np.array([6.4, 5.2, -8.1, -3.1])
vec_3 = np.array([-5.6, 6.4, 5.2, -8.1, -3.1, -2.0, 0.4, 9.4])
```

```python
def vecs_mult(vec1, vec2):
    counter_m = 0
    counter_n = 0
    len_vec1 = vec1.size
    len_vec2 = vec2.size
    mult_res = np.zeros(shape=(len_vec1, len_vec2))
    
    while counter_m < len_vec1 :
        while counter_n < len_vec2 :
            mult_res[counter_m][counter_n] = vec1[counter_m] * vec2[counter_n]
            counter_n = counter_n + 1
        counter_n = 0
        counter_m = counter_m + 1
    
    print(mult_res)
```

```python
vecs_mult(vec_3, vec_4)
```
