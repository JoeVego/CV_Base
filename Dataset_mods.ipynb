```python
result_dict = {}
# keys = ["labels", "boxes"]
# result_dict = dict([(key, []) for key in keys])
list_3 = []


for file in os.listdir(anno_path):
    filename = os.fsdecode(file)
    
    if filename[-3:] != "jpg":
        
        keys = ["labels", "boxes"]
        result_dict = dict([(key, []) for key in keys])
        
        with open(anno_path + "\\" + filename, 'r') as file:
            
            text = file.read()
    
            text = text.replace('\t','')
            text = text.replace('\n','')
            text = text.replace(',}','}')
            text = text.replace(',]',']')
            data = text.replace('0}{','0},{')
    
            file.close()
        
        data = data.replace("\'", "\"")
        data = "[" + data + "]"
        data_json = json.loads(data)
#         print(data_json)
        
        for element_of_UI in data_json:
#             print(element_of_UI)
            
            result_dict['labels'].append(element_of_UI.get("category_id"))
            result_dict['boxes'].append(element_of_UI.get("bbox"))
    
    
    pos_start = filename.find("_")
    
    new_name = filename[:pos_start] + "." + "txt"
    
    
    if new_name[0].isdigit() == True:
        
        with open(anno_path + "\\" + new_name, 'a') as output_file:
            json.dump(result_dict, output_file)
    
    
    result_dict = {}
```
