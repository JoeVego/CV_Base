```python
import torch
from torchvision import datasets
from torchvision import transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

https://pytorch.org/vision/stable/auto_examples/transforms/plot_transforms_illustrations.html#sphx-glr-auto-examples-transforms-plot-transforms-illustrations-py

```python
# trans = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])
trans = transforms.Compose([transforms.Resize(256),
                            # transforms.CenterCrop(200),
                            # transforms.RandomVerticalFlip(0.2),
                            # transforms.RandomPerspective(p = 0.9),
                            transforms.RandomAffine(80),
                            # transforms.ElasticTransform(),
                            transforms.RandomRotation(70),
                            # transforms.RandomCrop(100),
                            # transforms.RandomResizedCrop(150),
                            transforms.ColorJitter(brightness=.5, hue=.3),
                            transforms.ToTensor()])

train_dataset = datasets.MNIST(root = 'data/', train = True, transform=trans,download=True)
test_dataset = datasets.MNIST(root='data/', train=False, download=True)
```

```python
loader = DataLoader(dataset=train_dataset, batch_size=5,shuffle=True)
```

Функция permute() в PyTorch используется для перестановки размеров тензора в соответствии с указанным порядком.

Это может быть полезно в различных сценариях глубокого обучения, например, когда нужно изменить порядок измерений входных данных в соответствии с ожидаемым форматом ввода модели.
