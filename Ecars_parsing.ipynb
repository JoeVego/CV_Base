```python
import pandas as pd
from selenium import webdriver

from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Firefox()
```

```python
url = "https://www.drom.ru/catalog/~search/?wheel=0&fuel_type_short%5B0%5D=0&page=22"
```

```python
driver.get(url)
```

```python
links = driver.find_elements(By.XPATH, '//a[@class="b-link"]')
```

```python
list_links = []

for link_elem in links:
#     print(link_elem.get_attribute("href").count('/'))
    
    if link_elem.get_attribute("href").count('/') > 6 :
        list_links.append(link_elem.get_attribute("href"))
    

print(len(list_links))
```

```python
from selenium.common.exceptions import NoSuchElementException

list_capacity = []
list_power = []
list_name = []
list_errors = []
list_start_years = []

for link in list_links:
    try:
        driver.get(link)
    
        capacity = driver.find_element(By.XPATH, '//div[contains(text(), "Ёмкость")]/../*[2]')
        print(capacity.text)
        list_capacity.append(capacity.text)
    
        power = driver.find_element(By.XPATH, '//div[contains(text(), "Мощность")]/../*[2]')
        print(power.text)
        list_power.append(power.text)
    
        name = driver.find_element(By.XPATH, '//h1[contains(@class, "b-title b-title_type_h1")]')
        print(name.text[:-54])
        list_name.append(name.text[:-54])
        
        year = driver.find_element(By.XPATH, '//td[text()= "Период выпуска"]/../*[2]')
        print(year.text[-4:])
        list_start_years.append(year.text[-4:])
        
#         //td[text()= 'Период выпуска']/../*[2]
        
    except NoSuchElementException:
        name = driver.find_element(By.XPATH, '//h1[contains(@class, "b-title b-title_type_h1")]')
        list_errors.append(name.text)
        continue
    
pd1 = pd.Series(list_capacity)
pd2 = pd.Series(list_power)
pd3 = pd.Series(list_name)  
pd5 = pd.Series(list_start_years)

print(pd1)
print(pd2)
print(pd3)
print(pd5)
```

```python
pd4 = pd.Series(list_errors)
print(pd4)
```

```python
num_url = url[-2:]
excel_name = num_url + ".xlsx"

# создаем таблицу с текущим номером страницы
pd.concat([pd3, pd5, pd1, pd2], axis=1).to_excel(excel_name)
new_out = pd.read_excel(excel_name, index_col=0)


# читаем пред таблицу
num_url2 = int(num_url) - 1
excel_name2 = str(num_url2) + ".xlsx"

old_out = pd.read_excel(excel_name2, index_col=0)



# сохраняем новый файл объекдиненный
outs = [old_out, new_out]

pd.concat(outs).to_excel(excel_name)
print(excel_name)
```
