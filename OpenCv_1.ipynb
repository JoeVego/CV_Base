```python
def resizing():
    return cv2.resize(img, (1000, 600), cv2.INTER_NEAREST)
```

```python
plt.imshow(resizing())
```

```python
import numpy as np

def shifting():
    h, w = img.shape[:2]
    translation_matrix = np.float32([[1, 0, 200], [0, 1, 300]])
    dst = cv2.warpAffine(img, translation_matrix, (w, h))
    plt.imshow(dst)
```

```python
shifting()
```
